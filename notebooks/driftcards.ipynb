# Drift Cards

Drift cards were dropped in the Gulf of Mexico throughout 2013. This is on-going work to compare model results with the drift card retrieval locations.

Notes:

* Only driftcards that were dropped within the TXLA numerical domain were simulated with numerical drifters
* Only driftcards that were retrieved within the TXLA numerical domain were included in plots
* Direct windage has not been included yet
* The drifter simulations were run on hafen using aggregated hindcast OOF simulation output

```python
# Found info
ndrift = 321 #366 # number of found drift cards
enddates, endlons, endlats, endidnum = driftcards.readThenReduce(grid, ndrift, fileloc='../projects/driftcards/Recovery_data.txt')
# Deployed info, to connect idnums via start datetime
ndrift = 3490 # number of deployed drift cards
startdate, startlon, startlat, startidnum = driftcards.readThenReduce(grid, ndrift, fileloc='../projects/driftcards/Deployment_data.txt')
```

```python
# Loop through retrieval locations for drift cards and append together info for the ones that were found in the 
# domain and corresponding simulation tracks

Files = glob.glob('../tracks/driftcards/*.nc')

xp = []; yp = []; # simulation tracks
lonp = []; latp = []; # simulation tracks
x = []; y = [] # retrieval locations
ndays = [] # number of days drift card was at sea
# Loop through deployment simulations
for File in Files:
# File = Files[4]

    print File
    print 'reading...'

    d = netCDF.Dataset(File)
    simstarttime = d.variables['tp'][0]
    simstartlon = d.variables['lonp'][0,0]
    simstartlat = d.variables['latp'][0,0]
    lonptemp = d.variables['lonp'][:]
    latptemp = d.variables['latp'][:]
    tp = d.variables['tp'][:]
    d.close()
    
    # Find idnums associated with this simulation
    closetime=np.isclose(simstarttime,startdate)
    closelon=np.isclose(simstartlon,startlon)
    closelat=np.isclose(simstartlat,startlat)
    close = closetime*closelon*closelat
    ids = startidnum[close] # idnums that were represented in this simulation
        
    # Compare the starting idnums just found with the found drift card idnums
    iendidnum = find(np.in1d(endidnum,ids)) # indices of the retrieval entries that correspond to this simulation
    if not iendidnum.any():
        continue
    enddate = enddates[iendidnum] # the end dates for the corresponding retrievals
    endlon = endlons[iendidnum] # the end lon for the corresponding retrievals
    endlat = endlats[iendidnum] # the end lat for the corresponding retrievals
    
    ndays.append((enddate-simstarttime)/(3600.*24)) # number of days afloat for each drift card
    
    print 'found relevant retrieval locations for this simulation'
    
    # Find the index of time in simulation corresponding to the retrieval
    # for endd in enddate:
    #     itime = find(enddate[0]>tp)[-1]
    #     tracpy.plotting.tracks(lonp, latp, 'test', grid=grid)
    # tracpy.plotting.tracks(lonp[:,:itime], latp[:,:itime], 'test', grid=grid)
    # tracpy.plotting.transport(name='../tracks/driftcards', Title='Transport to Galveston', dmax=1.5)
    xptemp, yptemp = grid['basemap'](lonptemp, latptemp)
    ind = np.isnan(lonptemp)
    xptemp[ind] = np.nan; yptemp[ind] = np.nan
    xp.append(xptemp); yp.append(yptemp) # add on simulation tracks since a found drift card was found to correspond
    lonp.append(lonptemp); latp.append(latptemp) # add on simulation tracks since a found drift card was found to correspond
    
    xtemp, ytemp = grid['basemap'](endlon, endlat)
    x.append(xtemp); y.append(ytemp) # add on found locations for this set of cards/simulations
    # x, y = grid['basemap'](endlon[0], endlat[0])
    # Transport plot: can't align times, but can hope to show the retrieval is near transport pathway?
    # tracpy.plot
    
    print 'added info onto lists!'
    
```

This is a summary plot of all drifter tracks with the following specifics:

* Numerical drifters were started at the locations that driftcards were dropped that were within the TXLA numerical domain
* Numerical drifters were surface-limited
* Tracks are shown if at least one of the 10-30 drift cards dropped at a single location was retrieved within the numerical domain
* Tracks from a simulation are shown for up to the max number of days between dropping the drift cards and the last one retrieved. Some of the times are more than the 30 days that were simulation, in which case 30 days are shown
* Numerical drifters are I think started at the 4 hour mark before they were dropped, according to when model output was available

```python
days = (tp-tp[0])/(3600.*24)

fig = figure(figsize=(24,16))
tracpy.plotting.background(grid, fig=fig, mers=np.arange(-100,-86,2))

for i in xrange(len(x)):
    plot(x[i], y[i], 'sy', markersize=20, alpha=0.9)
for i in xrange(len(x)):
    if max(ndays[i]) <= 30:
        tind = find(days>=max(ndays[i]))[0]
        plot(xp[i][:,:tind].T, yp[i][:,:tind].T, color='0.4', linewidth=0.5, alpha=0.6) # tracks
    else: # show all available
        plot(xp[i][:].T, yp[i][:].T, color='0.4', linewidth=0.25, alpha=0.2) # tracks
# for i in xrange(len(x)):    
#     if max(ndays[i]) <= 30:
#         tind = find(days>=max(ndays[i]))[0]
#         xpc, ypc = tracpy.tools.find_final(xp[i][:,:tind], yp[i][:,:tind])
#         plot(xpc, ypc, 'ro', markersize=4, alpha=0.5) # ending points # ARE ANY EXITING THE DOMAIN?
#     else: # show all available
#         xpc, ypc = tracpy.tools.find_final(xp[i][:], yp[i][:])
#         plot(xpc, ypc, 'ro', markersize=4, alpha=0.5) # ending points # ARE ANY EXITING THE DOMAIN?
for i in xrange(len(x)):
    plot(xp[i][:,0], yp[i][:,0], 'go', markersize=10, alpha=0.8)
    
savefig('figures/no_final_positions_thinnerlightertracks.png', bbox_inches='tight')
```
