#All Forward: running drifters from through the TX-LA model domain forward in time
Kristen Thyng, September 2013

Numerical drifters are simulated from throughout the numerical model domain forward in time in 2006 for a month during the summer (July) and a month during the winter (February, to align with Zhaoru's shelf convergence work) for y days, and behavior is compared. Drifters are started once a day every 4 grid cells in x and y.

A parameterization of subgrid scale motion is implemented via the fluxes (`doturb=1`) and a horizontal diffusivity of $A_H=20$ m$^2$/s is used.

Results are not currently available online because pong is not currently NFS connected with hafen, but the [project](https://github.com/kthyng/gisr/blob/master/projects/all_f.py) is on github.

##Sample Results

Sample tracks and transport plots from a week-long simulation are shown below. Clearly, the majority of the transport is occurring on the outer shelf, due to loop current eddies. The transport highlights structures in the flow. Most of the plots show little cross-shelf transport, but there is some and an example was chosen from July 2006 to illustrate the behavior in the transport plot.

##Connectivity Maps
These are modeled after the NOAA Trajectory Analysis Planner (TAP) maps (["The NOAA Trajectory Analysis Planner: TAP II"](http://response.restoration.noaa.gov/sites/default/files/TAP_paper.pdf), Chris Barker, NOAA). They have a number of useful products all addressing pertinent response and response planning questions. I want to combine some aspects of several of these together.

###Galveston Bay Connectivity

Start by choosing a representative area near Galveston Bay. Drifters are tracked forward for 7 days. Add together all drifters that pass through the area. Note that currently, all drifters are treated as being equivalent, but potentially drifters could represent a different amount of initial volume transport or material. When a drifter is added into the Galveston Bay total, its initial location is aggregated on the map. The results are presented as a percentage of the total drifters from a location that pass through an area.

```python
# Load in drifter tracks
d = netCDF.MFDataset('tracks/all_f/2006-02*.nc', aggdim='ntrac')
lonpF = d.variables['lonp'][:]
latpF = d.variables['latp'][:]
d.close()
d = netCDF.MFDataset('tracks/all_f/2006-07*.nc', aggdim='ntrac')
lonpJ = d.variables['lonp'][:]
latpJ = d.variables['latp'][:]
d.close()

# number of files and therefore simulations
nfilesF = len(glob.glob('tracks/all_f/2006-02*.nc'))
nfilesJ = len(glob.glob('tracks/all_f/2006-07*.nc'))

# Check drifter tracks to see if they enter path and if they do, where they originated
sourceF = [] # list for original drifter locations that make it to Galveston Bay
for idrifter in xrange(lonpF.shape[0]): # loop through drifters
    # Find indices, ind, for drifter time steps that are near Galveston Bay
    indlon = find(lonpF[idrifter,:] < -94.0)
    indlat = find(latpF[idrifter,:] > 28.8)
    ind = list(set(indlon) & set(indlat))
    if ind == []:
        continue
    else:
        # Track drifter original location
        sourceF.append([lonpF[idrifter,0], latpF[idrifter,0]])
lonF = np.array(sourceF)[:,0]
latF = np.array(sourceF)[:,1]
sourceJ = [] # list for original drifter locations that make it to Galveston Bay
for idrifter in xrange(lonpJ.shape[0]): # loop through drifters
    # Find indices, ind, for drifter time steps that are near Galveston Bay
    indlon = find(lonpJ[idrifter,:] < -94.0)
    indlat = find(latpJ[idrifter,:] > 28.8)
    ind = list(set(indlon) & set(indlat))
    if ind == []:
        continue
    else:
        # Track drifter original location
        sourceJ.append([lonpJ[idrifter,0], latpJ[idrifter,0]])
lonJ = np.array(sourceJ)[:,0]
latJ = np.array(sourceJ)[:,1]
```

```python
fig = figure(figsize=(22,10))
# February Plot
ax = fig.add_subplot(1,2,1)
tracpy.plotting.hist(lonF, latF, 'all_f/galveston', grid=grid, which='hexbin', 
                    tind='other', C=1/(12.*nfiles), vmax=100., fig=fig, ax=ax, 
                    Title='February 2006', Label='Probability (%)')
xline, yline = grid['basemap']([-95.5,-94.2,-94.2],[28.8,28.8,29.6])
ax.plot(xline, yline, color='grey', linewidth=1)
# July Plot
ax = fig.add_subplot(1,2,2)
tracpy.plotting.hist(lonJ, latJ, 'all_f/galveston', grid=grid, which='hexbin', 
                    tind='other', C=1/(12.*nfiles), vmax=100., fig=fig, ax=ax, 
                    Title='July 2006', Label='Probability (%)')
ax.plot(xline, yline, color='grey', linewidth=1)
fig.suptitle('Probability of material making it to Galveston Bay in 7 days', fontsize=22, y=.94)
savefig('figures/all_f/galvestonhisthexbin.png', bbox_inches='tight')
```

```python
 

#ind = find(grid['h'] < 100.)
#for idrifter in xrange(lonp.shape[0]): # loop through drifters
# Interpolate inner value for drifter tracks
finner = grid['trirllrho'].nn_interpolator(inner.flatten())
innerpF = finner(lonpF, latpF)
innerpJ = finner(lonpJ, latpJ)
fouter = grid['trirllrho'].nn_interpolator(outer.flatten())
outerpF = fouter(lonpF, latpF)
outerpJ = fouter(lonpJ, latpJ)
```

###Shelf Connectivity

```python


sourceF = []
for idrifter in xrange(lonpF.shape[0]): # loop through drifters
    if (innerpF[idrifter,:] == 1).any() and (outerpF[idrifter,:] == 1).any():
        sourceF.append([lonpF[idrifter,0], latpF[idrifter,0]])
lonF = np.array(sourceF)[:,0]
latF = np.array(sourceF)[:,1]
    
sourceJ = []
for idrifter in xrange(lonpJ.shape[0]): # loop through drifters
    if (innerpJ[idrifter,:] == 1).any() and (outerpJ[idrifter,:] == 1).any():
        sourceJ.append([lonpJ[idrifter,0], latpJ[idrifter,0]])
lonJ = np.array(sourceJ)[:,0]
latJ = np.array(sourceJ)[:,1]
```

```python
fig = figure(figsize=(22,10))
# February Plot
ax = fig.add_subplot(1,2,1)
tracpy.plotting.hist(lonF, latF, 'all_f/shelf', grid=grid, which='hexbin', 
                    tind='other', C=1/(12.*nfiles), vmax=100., fig=fig, ax=ax, 
                    Title='February 2006', Label='Probability (%)')
ax.contour(grid['xr'], grid['yr'], grid['h'], [100], colors='grey', linewidth=2)

# July Plot
ax = fig.add_subplot(1,2,2)
tracpy.plotting.hist(lonJ, latJ, 'all_f/shelf', grid=grid, which='hexbin', 
                    tind='other', C=1/(12.*nfiles), vmax=100., fig=fig, ax=ax, 
                    Title='July 2006', Label='Probability (%)')
ax.contour(grid['xr'], grid['yr'], grid['h'], [100], colors='grey', linewidth=2)
fig.suptitle('Probability of material crossing the shelf in 7 days', fontsize=22, y=.94)
savefig('figures/all_f/shelfhisthexbin.png', bbox_inches='tight')
```
